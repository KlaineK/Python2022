#Nomor 2
Diketahui persamaan y'+8y = 2x^2 dan y(0)=2\
Tentukan y(0.2) dengan deret taylor orde 4\
Hitung estimasi error dan bandingkan dengan error sesungguhnya, dimakna solusi analitiknya\
image.png


```python
import math
```

```python
x = 0
y = 2
y_0order = 2
y_1order = (2*pow(x,2))-8*y
y_2order = 4*x - 16*pow(x,2) + 64*y
y_3order = 4 - 32*x + 64*(2*pow(x,2) - 8*y)
y_4order = -32 + 256*pow(x,2) - 512*(2*pow(x,2) - 8*y)

y_deriv = [y_0order, y_1order, y_2order, y_3order, y_4order]
h = 0.2

def taylor_series4() :
  taylor = 0
  for i in range(5):
    taylor += (y_deriv[i]*pow(h,i))/math.factorial(i)
  return taylor

print(f"Taylor Series Approximation : {taylor_series4()}")
```

```python
#Error Comparison
analytic_err = (31/32)*math.exp(-4*h) + 0.25*pow(h,2) - 0.125*h + 0.03125
abs_error = analytic_err - taylor_series4()

#Truncation Error
a = -32 + 256*pow(h,2) - 512*(2*pow(h,2) - 8*taylor_series4())
b = y_4order
trunc_err = (pow(h,4)/math.factorial(5))*(a-b)

print(f"Truncation Error : {trunc_err}\nAnalytic Error : {analytic_err}\nTrue Error : {abs_error}")
```
